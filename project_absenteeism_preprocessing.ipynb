Min, Max values are 0 and 28, indicating there are 29 numbers respectively. Whereas the length method shows that there are only 28 numbers in 'Reason for Absence'. And the missing number is 20.

As the numbers from 0-28(except 20) in 'Reason for Absence'column describes certain types of diseases or a reason for absence we will try to get the dummy variables that represents only 0 and 1, as this will make the dataset more readable.

Reasons source:
* https://clemence5.github.io/absenteeism/
* https://www.kaggle.com/datasets/tonypriyanka2913/employee-absenteeism

```python
reason_columns = pd.get_dummies(df['Reason for Absence'])
```

```python
reason_columns
```

```python
reason_columns['check'] = reason_columns.sum(axis = 1)
```

```python
reason_columns
```

```python
reason_columns['check'].sum(axis = 0)
```

```python
reason_columns['check'].unique()
```

```python
reason_columns = reason_columns.drop(['check'], axis = 1)
reason_columns
```

The index 0 column is not required, as from the 'Reasons of Absence' list there is no number significant to 0.

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
```

```python
reason_columns
```

### 2a) Grouping the columns 1-28 into a class

```python
df.columns.values
```

```python
reason_columns.columns.values
```

Drop the 'Reason for Absence' columns from 'df' table as the "reason_column" says conveys the same meaning. If not deleted, we might cause multicolinearity.

```python
df = df.drop(['Reason for Absence'], axis = 1)
```

```python
df
```

```python
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)
```

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis =1)
df
```

### 2b) Rearranging the column names

```python
df.columns.values
```

```python
#Changing the columns names
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
df.columns = column_names
```

```python
df.head()
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df = df[column_names_reordered]
```

```python
df.head()
```

### checkpoint 

```python
#modified and classified the "Reasons of Absence" column
df_reasons_modified = df.copy()
df_reasons_modified
```

# 3) Date

```python
#date is a string type
type(df_reasons_modified['Date'][0])
```

```python
df_reasons_modified['Date'] = pd.to_datetime(df_reasons_modified['Date'], format = '%d/%m/%Y')
```

```python
df_reasons_modified['Date']
```

```python
type(df_reasons_modified['Date'][0])
#Now the date column is converted to timestamp
```

```python
df_reasons_modified.info()
```

### 3a) Extracting the Month Values Only from df_reasons_modified

```python
df_reasons_modified['Date'][0].month
```

```python
#looping over df_reasons_modified, date columns to extraxct only the month values from it.
list_months = []
for i in range(df_reasons_modified.shape[0]):
    list_months.append(df_reasons_modified['Date'][i].month)
list_months
```

```python
len(list_months)
```

```python
#creating a new column named "Month Value" and appending it to df_reasons_modified
df_reasons_modified['Month Value'] = list_months
df_reasons_modified.head()
```

Month Values always ranges from 1-12.

### 3b)  Extracting the Day Of the Week Values Only from df_reasons_modified

```python
df_reasons_modified['Date'][699].weekday()
```

```python
#looping over df_reasons_modified, date columns to extraxct only the day of the week values from it.
list_days_of_week = []
for i in range(df_reasons_modified.shape[0]):
    list_days_of_week.append(df_reasons_modified['Date'][i].weekday())
list_days_of_week
```

```python
len(list_days_of_week)
```

```python
#creating a new column named "Day Of the Week" and appending it to df_reasons_modified
df_reasons_modified['Day Of the Week'] = list_days_of_week
df_reasons_modified.head()
```

### 3c) Dropping "Date" Column and rearranging the column names 

```python
df_reasons_modified.columns.values
```

```python
df_reasons_modified.drop(['Date'],axis = 1)
```

```python
#Rearranging the column names
columns_ordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value','Day Of the Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df_reasons_modified = df_reasons_modified[columns_ordered]
```

# 4) Education

```python
df_reason_date_modified['Education'].unique()
```

In "Education" the numbers 1, 2, 3, 4 represents as follows
* 1 : High School
* 2 : Graduate
* 3 : Postgraduate
* 4 : A master or doctorate

```python
df_reason_date_modified['Education'].value_counts()
```

Value 1 record showing that there are 583 high school greaduates alone, and over 110 candidates with  Graduate, Postgraduate, A master or doctorate degree combined. So it is an optimal way to combine these three records together

```python
df_reason_date_modified['Education'] = df_reason_date_modified['Education'].map({1:0 , 2:1, 3:1, 4:1})
```
